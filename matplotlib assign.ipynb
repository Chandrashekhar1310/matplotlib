# Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

Matplotlib is a Python data visualization library used to create 2D plots and graphs. It provides a way to represent data visually through a variety of chart types, such as line, scatter, bar, histogram, and pie charts.

Matplotlib is used to visualize data in a clear and concise way, which helps to understand complex data sets and trends. It is a popular choice among data analysts, scientists, and researchers due to its versatility and ease of use.

Here are five plots that can be plotted using the Pyplot module of Matplotlib:

1. Line plot: A line plot is a basic plot that displays data points connected by lines. It is used to visualize trends over time or to show the relationship between two variables.

2. Scatter plot: A scatter plot is used to display the relationship between two variables. It displays data as a collection of points, where each point represents a single observation.

3. Bar plot: A bar plot is used to compare data between different groups. It displays data as rectangular bars, where the height or length of the bar corresponds to the value of the data.

4. Histogram: A histogram is used to display the distribution of a set of data. It groups data into bins and displays the number of observations that fall into each bin.

5. Pie chart: A pie chart is used to display the proportion of different categories in a dataset. It is a circular chart that is divided into slices, where each slice represents a category and the size of the slice represents the proportion of data in that category.

# Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

A scatter plot is a type of graph that displays the relationship between two variables. It is also known as a scatter chart, scatter diagram, or scattergram.

In a scatter plot, each observation is represented as a point on the graph, where the x-axis represents one variable and the y-axis represents the other variable. The position of each point on the graph represents the values of both variables for that observation.

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y)
```

# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:

The subplot() function in Matplotlib is used to create multiple plots in a single figure. It is often used when we want to compare or visualize different aspects of the same data or related data sets.

```python
import numpy as np
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
plt.subplot(2, 2, 1)
plt.plot(x1, y1)
plt.title('Line 1')

plt.subplot(2, 2, 2)
plt.plot(x2, y2)
plt.title('Line 2')

plt.subplot(2, 2, 3)
plt.plot(x3, y3)
plt.title('Line 3')

plt.subplot(2, 2, 4)
plt.plot(x4, y4)
plt.title('Line 4')
```

# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

A bar plot is a type of plot in which rectangular bars are used to represent the data. The length or height of each bar is proportional to the value it represents. Bar plots are typically used to compare and visualize categorical data, where each category is represented by a bar.

Bar plots are commonly used because they are simple to create and easy to interpret. They allow us to easily compare the values of different categories, and can be used to show the distribution of a categorical variable.

For example, we can use a bar plot to show the sales of different products or the number of people in different age groups. Bar plots can also be used to compare the performance of different groups or teams.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company,profit)
plt.xlabel('Comapny')
plt.ylabel('profit')
```

```python
plt.barh(company,profit)
plt.xlabel('Profit')
plt.ylabel('Comapny')
```

# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

A box plot is a type of plot that is used to display the distribution of a dataset. It is also known as a box and whisker plot. The box plot displays the range of the data, as well as the median, quartiles, and outliers.

A box plot is made up of several components. The box represents the middle 50% of the data, with the bottom of the box being the first quartile (25th percentile) and the top of the box being the third quartile (75th percentile). The line inside the box represents the median (50th percentile) of the data. The whiskers extend from the box to show the range of the data, and outliers are plotted as individual points outside of the whiskers.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
data = [box1, box2]

fig, ax = plt.subplots()

ax.boxplot(data)

plt.title('Box Plot')
plt.xlabel('Data')
plt.ylabel('Values')

plt.show()
```
